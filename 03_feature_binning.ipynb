### 01. Import Dependecies

```python
import pandas as pd
```

### 02. Import Dataset

```python
df = pd.read_csv("data-process\cleaned_data.csv")
df.head(5)
```

### 03. Feature Binning

```python
def tenure_binning(tenure):
    if tenure <= 12:
        return 'Newer'
    elif tenure <=48:
        return 'Medium'
    else:
        return 'Long-term'
```

```python
df['tenure_bins'] = df['tenure'].apply(tenure_binning)
df.head(5)
```

```python
del df['tenure']
```
